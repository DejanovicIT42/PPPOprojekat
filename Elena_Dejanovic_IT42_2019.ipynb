# **2. Segmentacija**

Segmentacija je proces podele jedne slike na više segmenata (podgrupa) slike. Segment u tom slučaju predstavlja skup piksela na koje se deli slika, gde svaka regija sadrži piksele sa sličnim atributima. Dakle, segmentacija predstavlja postupak dodeljivanja oznaka pikselima, tako da svi elementi slike odnosno svi pikseli koji pripadaju istom segmentu imaju zajedničku oznako koja im je dodeljena jer dele određene osobine [1].


**2.1 Tehnike segmentacije**

Postoji nekoliko jedinstvenih prilaza segmentiranju slika.
Metode segmentacije slike traže objekte koji imaju neku meru homogenosti u sebi, ili imaju neku meru kontrasta na svojim granicama u odnosu na druge objekte. Većina algoritama za segmentaciju slike su modifikacije, proširenja ili kombinacije ova dva osnovna koncepta. Mere homogenosti i kontrasta mogu uključivati karakteristike kao što su nivo sive boje, ostalih boja i tekstura [1].
Najpopularinja podela na algoritme tehnika segmentacije:

1.   Segmentacija zasnovana na pragu (**Tresholding**)

  **Segmentacija pragom** predstavlja grupu metoda zasnovanih na poređenju osvetljenosti piksela sa jednim ili više pragova, pa tako postoji segmentacija sa jednim ili segmentacija sa više pragova. Ukoliko su u pitanju slike u boji tada se vrši poređenje boje između piksela. Segmentacija sa jednim pragom je najjednostavniji vid segmentacije i njegova osnovna primena je za odvajanje objekta od pozadine, ukoliko pozadina ima uniformnu osvetljenost (boju) koja se razlikuje od objekta. Ovakav način segmentacije se može primeniti, između ostalog, i kod izdvajanja nekog teksta, analize nekih medicinskih slika, prepoznavanja tipa aviona koji leti, itd. Segmentacija sa više pragova je metoda koju je pogodno koristiti u slučaju kada su u pitanju scene sa više različitih objekata [2].


2.   Segmentacija klasterizacijom (**Clustering**)

  **Klasterizacija (grupisanje)** je vrsta algoritma za mašinsko učenje bez nadzora i često se koristi za segmentaciju fotografija. Jedan od najdominantnijih algoritama zasnovanih na grupisanju koji se koristi za segmentaciju je *KMeans Clustering*. Ova vrsta grupisanja uglavnom koristi za pravljenje segmenata fotografija u boji, ali i sivih fotografija. Postupak segmentacije se u tom slučaju sastoji od podele N–dimenzionalnog prostora u uzajamno isključive oblasti, pri čemu svaka ta oblast obuhvata grupu podataka koja se odnosi na jedan region slike. Takav proces podele prema sličnosti vektora podataka se naziva klasterizacija. U prvoj fazi segmentacije se izračuna i izdvaja na obeležja. Zatim se određuje optimalan broj klastera, kao i skupa vektora koji predstavljaju centre svakog klastera. Nakon toga svaki piksel se pridružuje jednom od klastera. Pridruživanje se vrši na osnovu sličnosti vektora koji opisuje piksel i vektora koji opisuje centar klastera [2].
  
  Ova metoda je sa dosta uspeha primenjivana u segmentaciji multispektralnih satelitskih ili avionskih snimaka. Nedostatak ove metode je što ima veliku računsku složenost.

3.   Segmentacija regiona (**Region-based**)

  **Segmentacija regiona** predstavlja skup metoda koje pokušavaju da izdvoje područja slike koja su homogena sa gledišta određenih karakteristika [2]. U ovu grupu spadaju:

  1.   Segmentacija pomoću rasta regiona

      Osnovna ideja **segmentacije pomoću rasta regiona** je da se izvrši grupisanje susednih piksela sličnih osvetljenosti (boja), na osnovu čega se formiraju regioni. Metoda rasta regiona je naročito pogodna kod segmentacije prostih scena sa malim brojem objekata i slabom teksturom [2].

  2.   Segmentacija pomoću razdvajanja i spajanja regiona

      Kod **segmentacije pomoću razdvajanja i spajanja regiona**, slika se deli na četiri kvadranta, nakon čega se ispituje da li su dobijeni podsegmenti uniformni po obeležju na osnovu kojeg se vrši segmentacija [2].

  **Region** je grupa povezanih piksela koji pokazuju slična svojstva. Oni se mogu postaviti na primer, na osnovu intenziteta, boje i sl. U ovoj vrsti segmentacije prisutna su unapred definisana pravila koja piksel mora da ispuni kako bi bio klasifikovan u određeni region.

  Metode zasnovane na regionima su poželjnije od ostalih u slučaju kada se radi segmentacija slika sa smetnjama (kao što je to slučaj pri ultrazvuku) [3].

4.   Segmentacija granica (**Edge detection**)

  U procesu **segmentacije koji koristi ivice (granice regiona)** mora se primeniti i postupak za spajanje ivica, kako bi se formirale neprekidne granice regiona. Razlog za to je što metode koje vrše izdvajanje ivica generišu isprekidane granice objekata, a ne zatvorene krive. U tu svrhu se mogu koristiti: metoda spajanja ivica fitovanjem krive, spajanje ivica heurističkim metodama i spajanje ivica Hafovom transformacijom. Metodom fitovanja krivih se mogu rekonstruisati nedostajući segmenti ukoliko mapa ivica sadrži prekinute ivice. Fitovanje krivih je najjednostavnije kada nam je poznat oblik krive, pa se nedostajući segment dobija fitovanjem npr. pravih ili kružnih oblika. Kada su granice regiona složenije, vrši se razlaganje granice na jednostavnije delove. Glavna prednost ove metode je njena izuzetna jednostavnost i to da daje dobre rezultate prilikom segmentacije jednostavnih scena [2].
  
  Međutim, kod slika na kojima postoji veći broj objekata koji se preklapaju, mapa ivica sadrži tačke grananja, pa u tom slučaju ova metoda ne daje dobre rezultate. Kod heurističkih metoda za spajanje ivica se prvo formira gradijentna slika primenom odgovarajućih maski. Cilj segmentacije ivica je da dobijemo srednji rezultat segmentacije na koji možemo primeniti segmentaciju zasnovanu na regionu ili bilo koju drugu vrstu segmentacije da bismo dobili konačnu segmentiranu sliku [2].

Dodavanje funkcije ***save_checkpoint*** koja prati validation loss.
Ova funkcija sačuvaće checkpoint za model svaki put kada se validation loss smanji. Da bi ovo radilo neophodno je da imamo importovan Early stopping.



````python
def save_checkpoint(self, val_loss, model):
    import pickle
    if self.verbose:
        self.trace_func(f'Validation loss decreased ({self.val_loss_min:.6f} --> {val_loss:.6f}).  Saving model ...')
    with open(self.path, 'wb') as f:
        pickle.dump(model, f)
    self.val_loss_min = val_loss
````

Zatim definišemo funkciju ***create_csv*** koja će u .csv fajlu upisati slike i odgovarajuće maske za svaku sliku.

Varijabla:
```
 to_delete = f'_{key_word}'
```
će obrisati svaku sliku koja predstavlja masku.

Zatim funkcija vrši iteraciju kroz fajlove ***data_path***. Ukoliko naiđe na sliku koja je maska, funkcija će nastaviti da kreira red u .csv fajlu.
Ime datoteke slike se dobija uklanjanjem substringa to_delete iz naziva datoteke pomoću:
```
replace(to_delete, '')
```
Ovo se čuva u promenljivoj img.

Pandas DataFrame se kreira sa jednim redom, gde se ime datoteke slike stavlja u kolonu 'img', a ime datoteke maske u kolonu 'mask'. Na taj način smo u kodu jasno podelili slike iz foldera **train** na slike ultrazvuka i slike koje predstavljaju maske.

Ako izlazni .csv fajl naveden sa ***out_csv_path*** ne postoji, DataFrame se čuva kao novi .csv fajl koristeći:
```
data.to_csv(out_csv_path, header=True, indek=False)
```
Ako .csv fajl već postoji DataFrame se dodaje postojećem fajlu.






````python
def create_csv(data_path, out_csv_path, key_word='mask'):
    to_delete = f'_{key_word}'

    for file_name in os.listdir(data_path):
        if key_word in file_name:
            img = file_name.replace(to_delete, '')
            data = pd.DataFrame([img], index=['img']).transpose()
            data.insert(0, 'mask', file_name)

            if not os.path.exists(out_csv_path):
                data.to_csv(out_csv_path, header=True, index=False)
            else:
                data.to_csv(out_csv_path, mode='a', header=False, index=False)
````

Kreiranje seta podataka za model. U njega smeštamo te slike i odgovarajuće maske za trening:

````python
class ImageDataset(Dataset):
    def __init__(self, df, root_dir, transform=None):
        self.df = df
        self.root_dir = root_dir
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        mask = Image.open(os.path.join(self.root_dir, self.df.iloc[idx, 0]))
        image = Image.open(os.path.join(self.root_dir, self.df.iloc[idx, 1]))

        if self.transform:
            return self.transform(image), self.transform(mask)

        return image, mask
````

Definisanje i pozivanje funkcije ***draw_samples*** koja će prikazati nekoliko slika ultrazvuka i njihovih maski.

````python
def draw_samples(data, n_col, n_row):
    fig = plt.figure(figsize=(20, 5))

    for i in range(1, n_col + 1):
        img_ax = fig.add_subplot(n_row, n_col, i)
        msk_ax = fig.add_subplot(n_row, n_col, i + n_col)

        img_ax.imshow(data[i-1][0], cmap="gray")
        msk_ax.imshow(data[i-1][1], cmap="gray")

    fig.show()
````

**3.2. Model**

**Model** je deo koji se čuva nakon pokretanja algoritama nad podacima. Postoji lista podržanih kodera u SMP-u, uzima se odgovarajuća porodica enkodera i u okviru nje se bira konkretan enkoder koji dolazi sa unapred natreniranim podacima.

Postavljanje potrebnih parametara za ***segmentation_model_pytorch***:

````python
ENCODER = 'vgg11_bn'
ENCODER_WEIGHTS = 'imagenet'
ACTIVATION = 'sigmoid'
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
````

````python
model = smp.Unet(
    encoder_name=ENCODER,
    encoder_weights=ENCODER_WEIGHTS,
    in_channels=1,
    classes=1,
    activation=ACTIVATION
)
````

````python
loss = smpu.losses.DiceLoss()
metrics = [smp.utils.metrics.IoU()]
optimizer = torch.optim.Adam
scheduler = lr_scheduler.StepLR
````

Transformisanje slike

````python
my_transforms = transforms.Compose([
    transforms.Resize(size=(224, 224)),
    transforms.ToTensor() #transformiše iz PIL image u PyTorch FloatTensor
])
````

Podela data frame-a

````python
def split_df(df, fraction=0.8):
    df_1 = df.sample(frac=fraction)
    return df_1, df.drop(df_1.index)
````

**3.3. Treniranje**

Treniramo modela kako bi što uspešnije izvršio segmentaciju. Cilj je da utvrdimo dobre i prave vrednosti za sve tezine iz obeleženih i priloženih primera. Na ovaj način, model se gradi ispitivanjem različitih primera i pokušajem pronalska modela koji gleda da minimizira gubitak.  Poziva se ugrađena metoda za zaustavljanje treniranja u slučaju opadanja stope učenja, posle datog patience.



````python
def train(model, train_df, train_dir, optimizer, loss, metrics,
          learning_rate=0.01, batch_size=20, epochs=10, patience=3,
          scheduler=None, step_size=5, gamma=0.1, device='cpu', transform=None):

    early_stopping = EarlyStopping(patience, path='best_model.pkl', verbose=True)
    optimizer = optimizer(model.parameters(), learning_rate)

    if scheduler:
        scheduler = scheduler(optimizer, step_size, gamma)

    train_epoch = smp.utils.train.TrainEpoch(
        model, loss, metrics, optimizer, device, verbose=True
    )

    valid_epoch = smp.utils.train.ValidEpoch(
        model, loss, metrics, device, verbose=True
    )

    train_logs, valid_logs = [], []

    for epoch in range(epochs):
        train_dataframe, val_dataframe = split_df(train_df)
        train_dataset = ImageDataset(train_dataframe, train_dir, transform=transform)
        valid_dataset = ImageDataset(val_dataframe, train_dir, transform=transform)
        train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                                   batch_size=batch_size,
                                                   shuffle=True)
        valid_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                                   batch_size=batch_size,
                                                   shuffle=False)

        print(f'\nEpoch: {epoch+1}/{epochs}')

        train_log = train_epoch.run(train_loader)
        valid_log = valid_epoch.run(valid_loader)
        train_logs.append(train_log)
        valid_logs.append(valid_log)
        early_stopping(valid_log[loss.__name__], model)

        if early_stopping.early_stop:
            print("Early stopping")
            break

        if scheduler:
            scheduler.step()

    return train_logs, valid_logs
````

Proces treniranja po epohama:

````python
res = train(model=model,
            train_df=train_df,
            train_dir=train_path,
            optimizer=optimizer,
            loss=loss,
            learning_rate=0.01,
            metrics=metrics,
            batch_size=20,
            epochs=20,
            scheduler=scheduler,
            step_size=10,
            patience=3,
            device=DEVICE,
            transform=my_transforms)
````

**3.5. Testiranje**


Testiranje modela nad podacima iz test foldera, koji se potom čuvaju u submission fajlu.

````python
def rle_encoding(x):
    dots = np.where(x.T.flatten()==1)[0]
    run_lengths = []
    prev = -2
    for b in dots:
        if (b>prev+1): run_lengths.extend((b+1, 0))
        run_lengths[-1] += 1
        prev = b
    return run_lengths
````

````python
imgs = [f for f in os.listdir(test_path)]
imgs = sorted(imgs, key=lambda s: int(s.split('.')[0]))
````

Definisanje i pozivanje funkcije ***create_csv_submission*** koja će u .csv fajl upisati podatke o tome da li se na fotografiji nalazi nerv ili ne, u odnosu na prethodno istreniran model.

````python
def create_csv_submission(model, data_path, img_list, out_path):
    submission_df = pd.DataFrame(columns=['img', 'pixels'])
    model.to(DEVICE)
    model.eval()

    for i, img in enumerate(tqdm(img_list)):
        x = Image.open(os.path.join(data_path, img))

        x = my_transforms(x)

        x = x.unsqueeze(0).to(DEVICE)
        pred_mask = model.predict(x)

        pred_mask = pred_mask.cpu()
        pred_mask = transforms.Resize(size=(420, 580))(pred_mask)

        encoding = rle_encoding(pred_mask)

        pixels = ' '.join(map(str, encoding))
        submission_df.loc[i] = [str(i+1), pixels]

    submission_df.to_csv(out_path, index=False)
````

Pravi se submission fajl koji sadrži pretpostavke natreniranog modela:

````python
create_csv_submission(model=model,
                      data_path=test_path,
                      img_list=imgs,
                      out_path='submission.csv')
````
